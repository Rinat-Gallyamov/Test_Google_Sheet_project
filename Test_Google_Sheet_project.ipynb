```python
def qnt_leads(dim, p):
    if dim == 'd_utm_source':
        return (len(leads.lead_id[leads.d_utm_source == p]))
    elif dim == 'd_club':
        id_mans = club.manager_id[club.d_club == p]
        return (len(leads.lead_id[leads.l_manager_id.isin(id_mans)]))
    elif dim == 'd_manager':
        id_man = club.manager_id[club.d_manager == p]
        return (len(leads.lead_id[leads.l_manager_id.isin(id_man)]))
```

```python
qnt_leads('d_manager','manager #3')
```

```python
def qnt_trash_leads(dim, p):
    if dim == 'd_utm_source':
        return (len(leads.l_client_id[(leads.d_utm_source == p) & (leads.l_client_id == '00000000-0000-0000-0000-000000000000')]))
    elif dim == 'd_club':
        id_mans = club.manager_id[club.d_club == p]
        return (len(leads.l_client_id[(leads.l_manager_id.isin(id_mans)) & (leads.l_client_id == '00000000-0000-0000-0000-000000000000')]))
    elif dim == 'd_manager':
        id_man = club.manager_id[club.d_manager == p]
        return (len(leads.l_client_id[(leads.l_manager_id.isin(id_man)) & (leads.l_client_id == '00000000-0000-0000-0000-000000000000')]))
```

```python
def qnt_new_leads(dim, p):
    if dim == 'd_utm_source':
        return (len(leads.l_client_id[(leads.d_utm_source == p) & (leads.l_client_id.isin(new_cust_id))]))
    elif dim == 'd_club':
        id_mans = club.manager_id[club.d_club == p]
        return (len(leads.l_client_id[(leads.l_manager_id.isin(id_mans)) & (leads.l_client_id.isin(new_cust_id))]))
    elif dim == 'd_manager':
        id_man = club.manager_id[club.d_manager == p]
        return (len(leads.l_client_id[(leads.l_manager_id.isin(id_man)) & (leads.l_client_id.isin(new_cust_id))]))    
```

```python
def qnt_cust(dim, p):
    if dim == 'd_utm_source':
        return (len(result.l_client_id[(result.d_utm_source == p) & (result.distinction >= 0) 
                                     & (result.distinction <= 7)].drop_duplicates()))
    elif dim == 'd_club':
        id_mans = club.manager_id[club.d_club == p]
        return (len(result.l_client_id[(result.l_manager_id.isin(id_mans)) & (result.distinction >= 0) 
                                     & (result.distinction <= 7)].drop_duplicates()))
    elif dim == 'd_manager':
        id_man = club.manager_id[club.d_manager == p]
        return (len(result.l_client_id[(result.l_manager_id.isin(id_man)) & (result.distinction >= 0) 
                                     & (result.distinction <= 7)].drop_duplicates()))   
```

```python
def sum_inc(dim, p):
    n = result.l_client_id[result.distinction < 0]
    if dim == 'd_utm_source':
        return (result.m_real_amount[(result.d_utm_source == p) & (result.distinction >= 0)
                                                  & (result.distinction <= 7) & (result.l_client_id.isin(n) == False)].
                                                  sum())
    elif dim == 'd_club':
        id_mans = club.manager_id[club.d_club == p]
        return (result.m_real_amount[(result.l_manager_id.isin(id_mans)) & (result.distinction >= 0)
                                                  & (result.distinction <= 7) & (result.l_client_id.isin(n) == False)].
                                                  sum())
    elif dim == 'd_manager':
        id_man = club.manager_id[club.d_manager == p]
        return (result.m_real_amount[(result.l_manager_id.isin(id_man)) & (result.distinction >= 0)
                                                  & (result.distinction <= 7) & (result.l_client_id.isin(n) == False)].
                                                  sum())     
```

```python
def best_f(lst, func):
    res = []
    for i in range(1,len(lst)):
        if func == sum_inc:
            res.append(func(lst[0], lst[i]).item()) ##применяем item(), чтобы избежать Object of type int64 is not JSON serializable
        else:
            res.append(func(lst[0], lst[i]))
    return res
```
